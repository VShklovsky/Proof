```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

AmazonOrders_df = pd.read_csv('01-Jan-2009_to_27-Feb-2016.csv')
#print(AmazonOrders_df.columns)
#AmazonOrders_df
AmazonOrders_df[['Total Charged']] = (AmazonOrders_df['Total Charged'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))
#pd.to_datetime(AmazonOrders_df['Order Date'])
#AmazonOrders_df
Dates_df = AmazonOrders_df['Order Date' ]
Charges_df = AmazonOrders_df['Total Charged']

Dates_df_dt = pd.Series([pd.to_datetime(date) for date in Dates_df])

Charges_df_cs = Charges_df.cumsum()

plt.plot_date(x=Dates_df_dt, y=Charges_df_cs )
```
